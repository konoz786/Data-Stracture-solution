```python
import matplotlib.pyplot as plt
x=np.arange(-5,5,.25)
y=2*x**2+3*x+1
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y =2*x**2+3*x+1')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt

x=np.arange(-5,5,.5)
y=np.arange(-5,5,.5)
X,Y =np.meshgrid(x,y)
Z=1-X-Y
fig =plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Surface Plot of x + y + z = 1')
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
image = np.array(Image.open("Lenna_(test_image).png"))
image_size = image.shape
print("Image size:", image_size)
image_min = np.min(image)
image_max = np.max(image)
print("Minimum value:", image_min)
print("Maximum value:", image_max)
image_mean = np.mean(image)
image_std = np.std(image)
print("Mean:", image_mean)
print("Standard deviation:", image_std)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
resized_image = np.array(Image.fromarray(image).resize((70, 70)))
grayscale_image = np.array(Image.fromarray(resized_image).convert("L"))
plt.imshow(grayscale_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
stacked_vector = np.reshape(grayscale_image, (70*70, 1))
vector_size = stacked_vector.shape
print("Stacked vector size:", vector_size)
grayscale_image = Image.fromarray(grayscale_image)
grayscale_image.save("grayscale_image.png")
```
